```python
import numpy as np

import matplotlib.pyplot as plt

# Generar datos de x
x = np.linspace(0, 2*np.pi, 100)

# Calcular los valores de y
y = np.sin(x)

# Graficar la función seno
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Gráfico de la función seno')
plt.grid(True)
plt.show()
```
